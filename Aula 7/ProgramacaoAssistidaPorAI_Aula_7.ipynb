#### Passo 2 - Número Primo

```python
import math

def eh_primo(n):
    if n < 2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

# O Miniprojeto - Jogo da Forca
Um jogo da forca onde ao iniciar a execução, é gerada uma palavra aleatória e com isso é gerada também uma dica com relação à palavra gerada. Com a palavra gerada, é solicitado inputs do usuário pedindo uma letra para lançar tentativa de acerto a palavra. Havendo possibilidade de responder com a palavra inteira para confirmar acertos/erros. Não será possível acertar todas as letras para formar a palavra, haverá um limite para sobrar ao menos algumas letras faltando para então solicitar a palavra inteira ao usuário.

```python
import random

def jogar_forca():
    # Banco de palavras e dicas
    palavras_dicas = [
        ("PYTHON", "Linguagem de programação versátil"),
        ("ALGORITMO", "Sequência lógica de instruções"),
        ("COMPUTADOR", "Máquina para processamento de dados"),
        ("INTERNET", "Rede mundial de computadores"),
        ("VARIAVEL", "Espaço na memória para armazenar dados")
    ]

    # Escolha aleatória
    palavra, dica = random.choice(palavras_dicas)
    letras_descobertas = ["_" for _ in palavra]
    letras_tentadas = []

    # Limite para sobrar letras faltando (ex: não permite descobrir as últimas 2 letras via chute individual)
    LETRAS_RESTANTES_MINIMAS = 2

    print("=== BEM-VINDO AO JOGO DA FORCA ===")
    print(f"Dica: {dica}")

    while True:
        print(f"\nPalavra: {' '.join(letras_descobertas)}")
        print(f"Letras tentadas: {', '.join(letras_tentadas)}")

        # Verifica quantas letras faltam descobrir
        faltam = letras_descobertas.count("_")

        if faltam <= LETRAS_RESTANTES_MINIMAS:
            print("\nLimite de letras atingido! Agora você deve tentar adivinhar a palavra inteira.")
            chute_final = input("Qual é a palavra? ").strip().upper()
            if chute_final == palavra:
                print(f"Parabéns! Você acertou. A palavra era: {palavra}")
            else:
                print(f"Que pena! Você errou. A palavra correta era: {palavra}")
            break

        entrada = input("Digite uma letra ou a palavra inteira: ").strip().upper()

        if not entrada:
            continue

        # Se o usuário tentar a palavra inteira voluntariamente
        if len(entrada) > 1:
            if entrada == palavra:
                print(f"Incrível! Você acertou a palavra inteira: {palavra}")
            else:
                print(f"Erro fatal! A palavra não era '{entrada}'. Você perdeu. A correta era: {palavra}")
            break

        # Se o usuário tentar uma letra
        if entrada in letras_tentadas:
            print("Você já tentou essa letra!")
            continue

        letras_tentadas.append(entrada)

        if entrada in palavra:
            print(f"Boa! A letra '{entrada}' existe na palavra.")
            for i, letra in enumerate(palavra):
                if letra == entrada:
                    letras_descobertas[i] = entrada
        else:
            print(f"A letra '{entrada}' não existe na palavra.")

jogar_forca()
```
