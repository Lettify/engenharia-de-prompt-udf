#### Passo 2 - Número Primo

```python
import math

def eh_primo(n):
    if n < 2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

#### Passo 3 - Inversão de String

```python
def inverte_texto(texto):
    return texto[::-1]

print(inverte_texto("Olá"))
```
